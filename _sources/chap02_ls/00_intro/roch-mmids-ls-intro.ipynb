# Least squares

In this chapter and the next, we introduce two linear-algebraic workhorses of data science, least squares and singular value decomposition, and develop the mathematical concepts at their foundation.

This chapter has two main objectives: 

1. To introduce the various mathematical facets of the least squares problem: the algebra, the geometry, the numerics.

2. To discuss a key application: regression analysis.
