# Spectral graph theory

In this chapter, we look at network data.

This chapter has three main objectives: 

1. To define basic graph concepts.
2. To introduce ideas from spectral graph theory, in particular the eigenvalues of the Laplacian and variational characterization.
3. To discuss and implement some applications, including community detection and image segmentation.
