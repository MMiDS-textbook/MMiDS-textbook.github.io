# Singular value decomposition

In this chapter, we turn to the singular value decomposition.

This chapter has two main objectives: 

1. To introduce the singular value decomposition and how to compute it.
2. To discuss its many applications, including principle components analysis, low-rank approximation, pseudoinverses, and more.
